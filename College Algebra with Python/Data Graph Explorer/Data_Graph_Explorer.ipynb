```python
# First 2 columns to Numpy Array
column1_array = df.iloc[:, 0].to_numpy()
column2_array = df.iloc[:, 1].to_numpy()
```

```python
import matplotlib.pyplot as plt
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(column1_array, column2_array, color='blue', alpha=0.5)  # Assuming column1_array and column2_array contain the data
plt.title('Scatter Plot')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.grid(True)
plt.show()
```

```python
# Line graph
plt.figure(figsize=(8, 6))
plt.plot(column1_array, column2_array, color='red', marker='o', linestyle='-')  # Assuming column1_array and column2_array contain the data
plt.title('Line Graph')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.grid(True)
plt.show()
```

```python
import random
number_of_columns = len(df.columns.tolist())
random_columns = random.sample(range(number_of_columns), 2)

column1_array = df.iloc[:, random_columns[0]].to_numpy()
column2_array = df.iloc[:, random_columns[1]].to_numpy()

plt.figure(figsize=(8, 6))
plt.plot(column1_array, column2_array, color='red', marker='o', linestyle='-')  # Assuming column1_array and column2_array contain the data
plt.title('Line Graph')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.grid(True)
plt.show()
```
